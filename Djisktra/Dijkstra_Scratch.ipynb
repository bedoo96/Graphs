```python
nb_node = 1000
taille_Nom = 50
nb_colunm = 2
not_Found = -1
unknown = -1
infini = 999999
empty = " "
white = 0
black = 1
```

```python
# Define Node class
class Node:
    def __init__(self, name, father,weight,color):
        self.name = name
        self.father = father
        self.color = color
        self.weight = weight
        
    ### Add setters for the attributes ###
    def set_name(self, name):
        self.name = name
    def set_father(self, father):
        self.father = father
    def set_weight(self, weight):
        self.weight = weight
    def set_color(self, color):
        self.color = color

    ### Added getters for the attributes ###
    def get_name(self):
        return self.name
    def get_father(self):           
        return self.father
    def get_weight(self):
        return self.weight
    def get_color(self):
        return self.color

    def __str__(self):
        return f"Node ID: {self.id}, Name: {self.name}, Color: {self.color}, Parent: {self.parent}, Distance: {self.distance}"
```

```python
### Build the graph ###
class CGraph:
    def __init__(self):
        ### Add data to a file ###
        self.file = empty
        self.tab_node = [Node(empty, unknown, infini, white) for _ in range(nb_node + 1)]
        self.tab_edges = [[0 for _ in range(nb_node + 1)] for _ in range(nb_node + 1)]
        self.tab_edges_names = [[empty for _ in range(nb_node + 1)] for _ in range(nb_node + 1)]
        self.nb_node = 0

    ### Upload the table of nodes ###
    def upload_node(self):
        code_return = 0
        self.file = input("Enter the name of the Graph: ")
        nameFnodes = self.file + "_nodes.txt"
        try:
            with open(nameFnodes, "r") as file:
                for line in file:
                    line = line.strip()
                    if not line:
                        continue
                    name = line.strip().split()[0]
                    print(f"Reading node: '{name}'")  # DEBUG
                    if name != empty:
                        node = Node(name, unknown, infini, white)
                        self.tab_node[self.nb_node] = node
                        self.nb_node += 1
        except FileNotFoundError as e:
            print(nb_node, "nodes not found . Error: ", e.errno)
            code_return = not_Found
        return code_return


    def upload_edges(self):
        code_return = 0
        nameFedges = self.file + "_edges.txt"
        nameFnamesedges = self.file + "_name.txt"
        try:
            i = 0 
            with open(nameFedges, "r") as file:
                for line in file:
                    line = line.replace('\t', '')
                    if line.strip():
                        tab_edge = list(map(int, line.strip().split()))
                        self.tab_edges[i] = tab_edge
                        i += 1
            i = 0
            with open(nameFnamesedges, "r") as file:
                for line in file:
                    line = line.replace('\t', '')
                    if line.strip():
                        tab_edge = line.strip().split()
                        self.tab_edges_names[i] = tab_edge
                        i += 1

        except FileNotFoundError as e:
            print(nameFedges, "name Fedges not found . Error: ", e.errno)
            code_return = not_Found

        return code_return

    
    ### Node Choice ###
    def node_choice(self, texte):
        for i in range(self.nb_node):
            print("{0:2d} : ".format(i + 1), end=" ")
            print(" {:<50s}".format(self.tab_node[i].get_name()), end=" ")
            if (((i + 1) % nb_colunm) == 0):
                print()
                choice = int(input(("Enter the node number "+texte+" : ")))
                choice -= 1
                if (choice >= nb_node) or (choice < 0):
                    print(f"Wrong choice, please try again")
                    choice = unknown
        return choice
    
    ### Djikstra Algorithm Initialisation process ###

    def dijkstra_initialisation(self, start_node_number):
        self.tab_node[start_node_number].set_weight(0)


    ### Djikstra looking for minimal weight function ###
    def dijkstra_min_weight(self):
        num_node = not_Found
        min_weight = infini
        for i in range(self.nb_node):
            if self.tab_node[i].get_weight() < min_weight and self.tab_node[i].get_color() == white:
                min_weight = self.tab_node[i].get_weight()
                num_node = i
        return num_node
    
    ### Nodes color change function ###
    def dijkstra_color_change(self, node_number):
        self.tab_node[node_number].set_color(black)
    ### Weight of adjacent nodes function ###
    def dijkstra_weight_adjacent_nodes(self, node_number):
        for num_adjacent in range(self.nb_node):
                    new_weight = self.tab_node[node_number].get_weight() + self.tab_edges[node_number][num_adjacent]
                    if ((self.tab_node[num_adjacent].get_color() == white) and (new_weight< self.tab_node[num_adjacent].get_weight())):
                        self.tab_node[num_adjacent].set_weight(new_weight)
                        self.tab_node[num_adjacent].set_father(node_number)
                       
    ### Display the shortest path ###
    def dijkstra_display_shortest_path(self, end_node_number):
        nbsom = 1
        num_s = end_node_number
        tab_inverse_path = [0] * (nb_node + 1)

        if self.tab_node[end_node_number].get_weight() == infini:
            print("No path found")
        else:
            tab_inverse_path[0] = end_node_number
            while (num_s != unknown):
                tab_inverse_path[nbsom] = num_s
                nbsom += 1
                start_node = num_s
                num_s = self.tab_node[num_s].get_father()
            print("Shortest path to go from : ", end =" ")
            print(self.tab_node[start_node].get_name())
            print("to : ", end =" ")
            print(self.tab_node[end_node_number].get_name())
            i = nbsom - 1
            while (i > 0):
                num_s1= tab_inverse_path[i]
                if num_s1 != end_node_number:
                    print("-> That's the end", end = " ")
                    print(self.tab_node[num_s1].get_name(), end = " ")
                    print(" You have travalled from ", end = " ")
                    print(self.tab_node[end_node_number].get_weight(), end = " ")
                    print(" m")
                else:
                    num_s2 = tab_inverse_path[i - 1]
                    print("from ", self.tab_node[num_s1].get_name(), end = " ")
                    print("Follow ", self.tab_edges_names[num_s1][num_s2], end = " ")
                    print("to ", end = " ")
                    print(self.tab_node[num_s2].get_name(), end = " ")
                    print(" (", self.tab_edges[num_s1][num_s2], "m)")
                i -= 1
            print("That's the end")
                                
```

```python
### Main function ###

Graph = CGraph()

if Graph.upload_node() != not_Found:
    if Graph.upload_edges() != not_Found:
        message = " Start "
        start_node_number = Graph.node_choice(message)
        if start_node_number != not_Found :
            message = " End "
            end_node_number = Graph.node_choice(message)
            if end_node_number != not_Found:
                print("Dijkstra algorithm in progress...")
                Graph.dijkstra_initialisation(start_node_number)
                ### looking for advices of minimal weight ###
                num_node_min_weight = not_Found + 1
                while num_node_min_weight != not_Found:
                    num_node_min_weight = Graph.dijkstra_min_weight()
                    if num_node_min_weight != not_Found:
                        Graph.dijkstra_weight_adjacent_nodes(num_node_min_weight)
                        Graph.dijkstra_color_change(num_node_min_weight)
                Graph.dijkstra_display_shortest_path(end_node_number)
```
