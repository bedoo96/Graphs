## Graph Theory

<p> A graph is a mathematical structure consisting of a set of object, calledv <b>vertices or nodes </b> , and a set of connections, called <b>edges</b>
`G = (V, E)`

`G`: is the Graph

`V` : are the Nodes or vertices

`E` : Are the connections or set of edges

<h1> Undirected Graph</h1>

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

### Add vertices and edges ###
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B','E'), ('C', 'F'), ('C', 'G')])

### Display informations about G ###
print (G)

### plot the Graph G ###
plt.axis('off')
plt.title("Undirected Graph")
nx.draw_networkx(G,
                 pos = nx.spring_layout(G, seed= 0),
                 node_size = 600,
                 cmap = 'coolwarm',
                 font_size = 14,
                 font_color = 'white')
```

<h1>Directed Graph or Digraph</h1>

```python
### create a Digraph object ###
DG = nx.DiGraph()

### Add vertices and edges ###
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B','E'), ('C', 'F'), ('C', 'G')])

### plot the Digraph ###
plt.axis ("off")
plt.title(" Digraph")
nx.draw_networkx(
    DG,
    pos = nx.spring_layout(DG, seed = 0),
    node_size = 600,
    cmap = "coolwarm",
    font_size  = 14,
    font_color = "White"
)
```

<h1>Weigthed Graph</h1>

<p>In a weighted graph each edge has a weight or a cost associeted with it, in the following case (Transportation), the weight can be considered as travel time or distance between cities</p>

```python
### let's add weight to G ###

WG = nx.Graph()

### Add vertices , edges and weight ###
WG.add_edges_from ([('A', 'B' , {"weigth":10}), ('A', 'C', {"weight" : 20}), ('B', 'D', {"weight":30}),
                    ('B','E', {"weight": 40}), ('C', 'F', {"weight":50}), ('C', 'G', {"weight": 60})])

labels = nx.get_edge_attributes(WG, "weight")

### plot the Digraph ###
plt.axis ("off")
plt.title(" Weighted Undirected Graph")
nx.draw_networkx(
    WG,
    pos = nx.spring_layout(G, seed = 0),
    node_size = 600,
    cmap = "coolwarm",
    font_size  = 14,
    font_color = "White"
)
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(G, seed =0), edge_labels=labels)
```

<h1>Connected graph</h1>

```python
### Create a graph ###
G1 = nx.Graph ()

### Add vertices and edges 
G1.add_edges_from([(1, 2), (2, 3), (3, 1), (4, 5)])
print(f"Is a graph 1 connected? {nx.is_connected(G1)}")
nx.draw_networkx(
    G1,
    pos = nx.spring_layout(G1, seed = 0),
    node_size = 600,
    cmap = "coolwarm",
    font_size  = 14,
    font_color = "White"
)

### create a second graph ###
G2 = nx.Graph()

### Add vertices and edges ###
G2.add_edges_from([(1, 2), (2, 3), (3, 1), (1, 4)])
print(f"Is graph 2 is connected? {nx.is_connected(G2)}")
nx.draw_networkx(
    G2,
    pos = nx.spring_layout(G2, seed = 0),
    node_size = 600,
    cmap = "coolwarm",
    font_size  = 14,
    font_color = "White"
)
```

<h1>Degree of Node</h1>

<h4> Calculate the node degree, indegree, or outdegree using built-in methods</h4>

```python
# calculate degree on Node

### Create a instance of graph Undirected ###
G = nx.Graph()
### Add vertices and edges to G ###
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 
'E'), ('C', 'F'), ('C', 'G')])
### print degrees ###
print(f"deg(A) = {G.degree['A']}")

### Other Graph Directed###
DG = nx.DiGraph()
### Add vertices and edges to G ###
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 
'E'), ('C', 'F'), ('C', 'G')])
### print degrees ###
print(f"deg^-(A) = {DG.in_degree['A']}")
print(f"deg^+(A) = {DG.out_degree['A']}")
```

```python
print(f"Degree centrality = {nx.degree_centrality(G)}")
print(f"Closeness centrality   = {nx.closeness_centrality(G)}")
print(f"Betweenness centrality = {nx.betweenness_centrality(G)}")
```
